```python
import pandas as pd
```

```python
air_quality_url = "https://raw.githubusercontent.com/konrad/DaLI_Basismodul_WiSe2021_2022/main/data/Rodenkirchen_processed.csv"
```

```python
# Wir lesen unsere Date in den DateFrame ein
air_quality = pd.read_csv(air_quality_url, na_values=["-"], index_col=0)
```

```python
air_quality
```

```python
# Wir generieren eine Spalte mit sauberen Datumsstrings
air_quality["Datum_sauber"] = air_quality["Datum"].apply(
    lambda date_string: 
        date_string[7:11] + "-" + date_string[4:6] + "-" + date_string[1:3])
```

```python
# Wir generieren eine Spalte mit sauberen Uhrzeitstrins
air_quality["Uhrzeit"] = air_quality["Datum"].apply(
    lambda date_string: 
        date_string[12:-1])
```

```python
air_quality
```

```python
# Wir nehmen die Werte um 3 Nachmittags als Wert für den Tag
air_quality_day = air_quality[air_quality["Uhrzeit"] == "15:00"]

# Wir nehmen die Werte um 3 Nacht als Wert für die Nacht
air_quality_night = air_quality[air_quality["Uhrzeit"] == "03:00"]
```

```python
air_quality_night
```

```python
# Wir verwerfen alle nicht benötigten Spalten der Übersichtlichkeit halber
air_quality_day = air_quality_day[["Datum_sauber", "Feinstaub (PM₁₀) stündlich gleitendes Tagesmittel in µg/m³"]]
air_quality_night = air_quality_night[["Datum_sauber", "Feinstaub (PM₁₀) stündlich gleitendes Tagesmittel in µg/m³"]]
```

```python
air_quality_day
```

```python
# Wir führen die Feinstaubwert-DataFrames eine Tages 
# für die Tag- und die Nachtmessung zusammen.
# Da die Feinstaubspalte in beiden DataFrames die gleiche Bezeichnung hat,
# hängt pandas ein "_x" bzw. ein "_y" an den Namen.
air_quality_day_vs_night = air_quality_day.merge(air_quality_night, on="Datum_sauber")
```

```python
air_quality_day_vs_night
```

```python
# Wir vergleichen die Feinstaubwerte der verschiedenen Tageszeiten für jeden Tag
air_quality_day_vs_night["Wert_tagsueber_groesser_als_nachts"] = air_quality_day_vs_night[
    "Feinstaub (PM₁₀) stündlich gleitendes Tagesmittel in µg/m³_x"] > air_quality_day_vs_night[
    "Feinstaub (PM₁₀) stündlich gleitendes Tagesmittel in µg/m³_y"]
```

```python
air_quality_day_vs_night
```

```python
air_quality_day_vs_night.rename(columns={
    "Feinstaub (PM₁₀) stündlich gleitendes Tagesmittel in µg/m³_x": "Feinstaub Tag in µg/m³",
    "Feinstaub (PM₁₀) stündlich gleitendes Tagesmittel in µg/m³_y": "Feinstaub Nacht in µg/m³"
})
```
