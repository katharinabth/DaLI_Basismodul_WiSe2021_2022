## Bearbeiten und zusammenführen der beiden Datensets


Wir werden hier die beiden Datensets, die in vorherigen Veranstaltungen besprochen wurden bearbeiten. Diese stammen von 
- https://www.umweltbundesamt.de/daten/luft/luftdaten/luftqualitaet
- https://meteostat.net/de/place/45I0JU?t=2020-11-01/2021-10-31


Die Daten wurden um den Einstieg etwas zu vereinfachen etwas aufbereitet und sind hier (mit diesen Jupyter Notebook) zu finden:

- https://github.com/konrad/DaLI_Basismodul_WiSe2021_2022

```python
# Packages in das Jupyter Notebook laden
import urllib.request
import pandas as pd
```

```python
# Setzen der URL-Variablen
air_quality_url = "https://raw.githubusercontent.com/konrad/DaLI_Basismodul_WiSe2021_2022/main/data/air_qualities_cologne.csv"
weather_url = "https://raw.githubusercontent.com/konrad/DaLI_Basismodul_WiSe2021_2022/main/data/weather_cologne.csv"

# Setzen von Datein-Namen-Variablen
air_quality_file = "air_quality_cologne.csv"
weather_file = "weather_cologne.csv"
```

```python
# Mittels der Funktion urlretrieve werden die Daten für die Luftqualität heruntergeladen
urllib.request.urlretrieve(air_quality_url, air_quality_file)
```

```python
# Mittels der Funktion urlretrieve werden die Wetter-Daten heruntergeladen
urllib.request.urlretrieve(weather_url, weather_file)
```

```python
# Die Daten werden in einen "DataFrame" von pandas eingelesen
# Eine gut Einfürhung ist hier zu finden:
# https://pandas.pydata.org/docs/getting_started/index.html#getting-started
weather = pd.read_csv(weather_file)
```

```python
# Den Wetter-DataFrame anzeigen lassen
weather
```

```python
# Die Methodes "describe" gibt ein Überblicksstatistiken für numerische Spalten zurück
weather.describe()
```

```python
# Die Daten zur Luftqualität in einen DataFrame einlesen
air_quality = pd.read_csv(air_quality_file)
```

```python
air_quality
```

```python
# Die Methodes "describe" gibt ein Überblicksstatistiken für numerische Spalten zurück
# ... in diesem Fall sind die Spalten allerdings nicht nummerisch, das fehlende Daten durch
# "-" gekennzeichnet wurden und damit die Spalte von pandas als String-Werte gehandelt wird.
air_quality.describe()
```

```python
air_quality["Feinstaub (PM₁₀) stündlich gleitendes Tagesmittel in µg/m³"].unique()
```

```python
# Wir lassen die Daten noch einmal neu Einlesen. Diesmal sagen wir das Zellen mit "-" den
# Wert "NA" (not available/not applicable) bekommen. Pandas kann die Spalten dann als Zahlenspalte
# interpretieren.
air_quality = pd.read_csv(air_quality_file, na_values=["-"])
```

```python
air_quality.dtypes
```

```python
# Jetzt zeigt "describe" schön Übersichtsstatistiken.
air_quality.describe()
```

```python
# Wir können die Daten an Hand der jeweiligen Spalte mit dem Datum zusammenführen.
air_quality_and_weather = air_quality.merge(weather, left_on="Datum", right_on="date")
```

```python
air_quality_and_weather
```

```python
# Wir können uns mit der Methode "hist" ein Histogram für eine Spalte erstellen lassen.
air_quality_and_weather.hist("Feinstaub (PM₁₀) stündlich gleitendes Tagesmittel in µg/m³")
```

```python
# Wir können stat der standarmäßigen 10 bin auch mehr bins nutzen und die Farbe
# Histrogramms ändern.
air_quality_and_weather.hist("Feinstaub (PM₁₀) stündlich gleitendes Tagesmittel in µg/m³", bins=30, color="pink")
```

```python
# Um einen neuen DataFrame mit einer ausgewählten Spaltenauswahl zu erzeugen,
# nehmen wir eine Liste der Spaltennamen als Index.
air_quality_and_weather_selection = air_quality_and_weather[
    ["Stationscode", "Datum", "Uhrzeit", "Luftqualitätsindex"]]
```

```python
air_quality_and_weather_selection
```

```python
# Wir können Zeilen, die eine bestimmten Wert haben herausfiltern.
air_quality_and_weather[air_quality_and_weather["Luftqualitätsindex"] == "sehr gut"]
```

```python
# Wir können Zeilen, die eine bestimmten Wert überschreiten herausfiltern.
air_quality_and_weather[air_quality_and_weather[
    "Feinstaub (PM₁₀) stündlich gleitendes Tagesmittel in µg/m³"] > 50.0]
```
