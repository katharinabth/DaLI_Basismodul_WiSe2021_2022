## Luftqualitätsdaten

````python
import pandas as pd
import numpy as np
````

````python
url = "https://www.umweltbundesamt.de/api/air_data/v2/airquality/csv?date_from=2020-11-01&time_from=1&date_to=2021-11-01&time_to=1&station=1254&lang=de"
````

````python
air_quality = pd.read_csv(url, sep=";", skipfooter=1, engine="python")
````

````python
air_quality["date_nomalized"] = air_quality["Datum"].apply(
    lambda date: f"{date.split('.')[2].split()[0]}-{date.split('.')[1]}-{date[1:].split('.')[0]}")
````

````python
air_quality
````

````python
combined = air_quality.merge(weather, left_on="date_nomalized", right_on="date")
````

````python
combined
````

````python
combined = combined[["Feinstaub (PM₁₀) stündlich gleitendes Tagesmittel in µg/m³", "tmax", "Luftqualitätsindex"]]
````
